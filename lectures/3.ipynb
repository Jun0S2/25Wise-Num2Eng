## Step 2:

A common time stepping scheme to solve our system of ODEs implicit Trapezoidal rule

For a scalar ODE :

$$
\dot y(t) = f(t,y(t))
$$

it is given by $y^{n+1} = y^n + \frac{\Delta t}{2} [f(t_n, y^n)+f(t_{n+1},y^{n+1})]$ 

- $y^{n+1}$ : unknown

$$
F(y) = -y + y^n+ \frac{\Delta t}{2}[f(t_{n},y^{n})+f(t_{n+1},y)] ]
$$

Find y such that F(y) = 0 (이럴떈 뉴턴의 메서드를 쓴데 머고이게)

<aside>

$$
f \to A U + \frac{1}{h^2}g
$$

</aside>

$$
U^{n+1}=U^n+\frac{\Delta t}{2}[AU^n+\frac{1}{h^2}g(t^n)+AU^{n+1}+\frac{1}{h^2}g(t^{n+n})]
$$

- $AU^{n+1}$ : unkown
- $\frac{1}{h^2}g(t^{n+n})$ : unkown

→ This theme is called

<aside>

# Crank Nicolson

$$
\to (I - \frac{\Delta t}{2}A)U^{n+1} = U^n+\frac{\Delta t}{2}[AU^n+\frac{1}{h^2}(g(t_n) + g(t_{n+1}))]
$$

</aside>

To update $U^n \to U^{n+1}$ we need to solve a linear system with $(I - \frac{\Delta t}{2}A)$

- I : identity matrix
- $(I - \frac{\Delta t}{2}A)$ : tridiagonal shape이래 왜용
- so its cheap to solve this system 왜용
