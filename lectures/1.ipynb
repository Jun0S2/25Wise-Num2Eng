### What happens if h→0?

**Example**

**what does it mean?**

$$
f(h) = 2h^2 + 8h^5
$$

- h = 0.1 : $h^2 = -0.02, h^5 = 0.0008$
- h = 0.01 : $2h^2 = 0.0002, 8h^5=0.0000000008$

it will go much faster to 0 if we make h much smaller

As `h->0`  the behavior of f is dominated by $2h^2$

**so what can we say about this example?**

- $f(h) = 2h^2+O(h^5)$, (k=8) …. since $Kh^p$
- $f(h) = O(h^2)$ … big O notation
    - formally, $h_0 = 0.1$, $2h_0^2 > 8h_0^5 => K=4$
    - $|2h^2 + 8h^5|$ ≤ $Kh^2$
    - K =2 해도 됌. 대체로 가장 작은 지수의 상수로 선택해도 됌.
</aside>

## Definition

We say that 

$$
g(h) = O(h^p)
$$

- for `h->0`

if there exists a positve number K (independent of h), such that $|g(h)| <= Kh^p$ for h→0 (h≤ $h_0$)

**In words** : $g(h)$ goes to zero (at least) as fast as $Kh^p$  does

---

<aside>

**Big O** allows us to focus on the leading order (error) term

</aside>

## Remark

For $x$ → $\infin$ : $f(x) = O(g(x))$ as $x$ → $\infin$ (stands for “f(x) is the order of g(x)”),

if there exists a positive real number K and a real number $x_0$ such that

$$
|f(x)| \le K|g(x)|
$$

- for all x = x0
