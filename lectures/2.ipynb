# 2.2 FD Scheme for linear advection Equation

- $u(0,x) = u_0(x) \forall x \in [a,b]$,
- $u(t,a) = u(t,b) \forall t \in [0,T]$

# FTCS

$$
u_t+au_x = 0
$$

- to approximate $u_x$ : use central diffquo $D_1^c$

ux가 first derivativ냐? 왜 central 을 쓰냐? 2nd order라서 더 정확하대 이게 다 먼솔

$$
u_(.,x_k) == \frac{u(.,x_{k+1}) - u(.,x_{k-1})}{2h} = \frac{U_{k+1}-U_{k-1}}{2h}
$$

- to approximate $u_t$ : forward diffequo (explicit in time)

$$
u_t (t^n,.) == \frac{u(t^{n+1},.)-u(t^n,.)}{\Delta t} \to \frac{U^{n+1}-U^n}{\Delta t}
$$

$$
 \to \frac{U_k^{n+1}-U_k^n}{\Delta t} + a \frac{U_{k+1}^{n}-U_{k-1}^n}{2h} =0
$$

# Upwind scheme

- a>0 : $U_k^{n+1} = U^n_k - \frac{a\Delta t}{h} (U^n_k - U^n_{k-1})$
- a<0 : $U_k^{n+1} = U^n_k - \frac{a\Delta t}{h} (U^n_{k+1} - U^n_{k})$
- is a first order scheme
- is stable under the time step constraint (CFL conditional)

$$
\Delta t \le \frac{h}{|a|}
$$

the biggert the speed, 말을 하다말ㅇ아..

We implement that :

- $\Delta t = \nu \frac{h}{|a|}, \nu \in R_+$

v 가 아니라 뉴f라고 해.

Stable for 0≤ v ≤1. CFL coeff / number

In practice : v = 0.8, 0.9 를 사용한다.
